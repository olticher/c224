1. Напишите регулярное выражение, которое возвращает список первых двух букв каждого слова строки. Обратите внимание на работу с дефисом.

```
import re
```

```
text = 'Здравствуйте, Анастасия Вадимовна! Надеюсь, вам понравится моя лаба, ведь я-то очень старалась.'
re.findall(r'\b[а-яA-Яa-zA-Z]{2}', text)
```

2. Напишите регулярное выражение, которое выбирает из строки все слова, в которых
строго больше 3 символов.

```
re.findall(r'[а-яA-яa-zA-Z]{4,}', text)
```

3. Напишите регулярное выражение, которое заменит все подстроки, обозначающие
время (только время, не даты), в строке на TBD.


```
text = 'Сейчас 22:59:04, немного хочется спать. В 0:30 постараюсь лечь. Кстати, мне 20 лет.'
re.sub(r'(\d{0,2}):(\d{2}:?)+', 'TBD', text)
```

4. Напишите регулярное выражение, которое заменяет произвольное количество пробельных символов внутри строки на один пробел.

```
text = 'Сейчас 22:59:04      , немного   хочется спать. В 0:30        постараюсь лечь. Кстати,  мне 20 лет .'
re.sub(r'\s+', ' ', text)
```

5. Напишите регулярное выражение, которое удаляет идущие подряд повторы. Одно
слово из группы должно остаться.

```
text = 'Как не не не не плакать по тебе неблагодарных слез слез? Но они опять опять текут, хоть я их их не стыжусь стыжусь'
re.sub(r'\b([^\W]+)(\s+\1)+\b', r'\1', text)
```

6. Напишите регулярное выражение, которое определяет, что подстрока является адресом электронной почты.

```
text = 'Моя почта: plotskaja2001@yandex.ru. Не почта в принципе: plotskaja@2001@yandexru.'
mail = re.search(r'\S+@\w+\.\w+', text)
mail.group()
```

7. Напишите регулярное выражение, которое возвращает список аббревиатур в строке.

```
text = 'ИТМО вроде как уже не аббревиатура, а вот МГУ все еще да. Кстати, я не люблю BTS.'
re.findall(r'[А-ЯA-Z]{2,}', text)
```

8. Напишите регулярное выражение, которое разделяет текст на предложения.

```
text = 'Первое предложение... А это второе? Да, а вот это третье!'
re.split(r'(?<=\w[.?!]\s)', text)
```

9. Напишите регулярное выражение, которое определяет, что строка является номером российского мобильного телефона любого оператора.

```
text = '89116735482, +79562345678, +17862346578, 491123456789'
re.findall(r'(?:\+7|8)\d{10}', text)
```

10. Напишите регулярное выражение, которое проверяет, что все предложения в строке начинаются с заглавной буквы.

```
text = 'Первое предложение. А это второе? да, а вот это третье!'
sentenses_number = len(re.split(r'(?<=\w[.?!]\s)', text))
upper_number = len(re.findall(r'(?:[.!?]\s[A-ZА-Я])|(?:^[A-ZА-Я])', text))
if sentenses_number == upper_number:
  print('Все предложения в строке начинаются с заглавной буквы')
else:
  print('Перепроверь текст. Где-то не хватает заглавной буквы!')
```

## 2.2 Изучите собранный вами в работе 1 корпус текстов и примените регулярные выражения для очистки от лишней информации (тегов html-разметки, url, непонятных символов и др.) или приведения отдельных потенциально важных элементов текстов к единому виду (времени или дат, например).
Сохраните обновленный корпус в формате json.

```
import json
```

```
with open("Severodvinsk.json") as f:
    data = json.load(f)
```

```
listik = list(data.values())[:11]
listik
```

```
corpus = str(list(data.values())[0])
for i in range(1, len(data)-1):
    corpus += ' '
    corpus += list(data.values())[i]
```

Что надо удалить в корпусе:

1. \n (переход на новую строку)
2. Все, что находится в обычных и фигурных скобках
3. Буквы под ударением (заменим их на обычные)
4. Знак ==





 

1. Уберем символ перехода на новую строку

```
corrected_corpus = re.sub(r'\n', ' ', corpus)
```

```
corrected_corpus
```

2. Уберем скобки и все, что в них

```
corrected_corpus = re.sub(r'\([^()]*\)', '', corrected_corpus)
corrected_corpus = re.sub(r'\([^{}]*\)', '', corrected_corpus)
```

```
corrected_corpus
```

3. Заменим гласные под ударением на обычные

```
corrected_corpus = re.sub(r'а́', 'а', corrected_corpus)
corrected_corpus = re.sub(r'А́', 'А', corrected_corpus)
corrected_corpus = re.sub(r'е́', 'е', corrected_corpus)
corrected_corpus = re.sub(r'É', 'Е', corrected_corpus)
corrected_corpus = re.sub(r'и́', 'и', corrected_corpus)
corrected_corpus = re.sub(r'И́', 'И', corrected_corpus)
corrected_corpus = re.sub(r'о́', 'о', corrected_corpus)
corrected_corpus = re.sub(r'О́', 'О', corrected_corpus)
corrected_corpus = re.sub(r'у́', 'у', corrected_corpus)
corrected_corpus = re.sub(r'У́', 'У', corrected_corpus)
corrected_corpus = re.sub(r'ы́', 'ы', corrected_corpus)
corrected_corpus = re.sub(r'Ы́', 'Ы', corrected_corpus)
corrected_corpus = re.sub(r'ю́', 'ю', corrected_corpus)
corrected_corpus = re.sub(r'Ю́', 'Ю', corrected_corpus)
corrected_corpus = re.sub(r'э́', 'э', corrected_corpus)
corrected_corpus = re.sub(r'Э́', 'Э', corrected_corpus)
corrected_corpus = re.sub(r'я́', 'я', corrected_corpus)
corrected_corpus = re.sub(r'Я́', 'Я', corrected_corpus)
```

```
corrected_corpus
```

4. Уберем знак ==

```
corrected_corpus = re.sub(r'==', '', corrected_corpus)
```

```
corrected_corpus
```

```
with open('Severodvinsk.json', 'w') as f:
    json.dump(corrected_corpus, f, ensure_ascii=False)
```
